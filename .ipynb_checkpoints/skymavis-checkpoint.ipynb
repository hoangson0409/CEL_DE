with 
transformed_items1 as (
select *, REGEXP_SUBSTR(cost, '[0-9]+(?=\\s+gold)') AS gold_cost,
REGEXP_SUBSTR(cost, '[0-9]+(?=\\s+silver)') AS silver_cost
REGEXP_SUBSTR(cost, '[0-9]+(?=\\s+bronze)') AS bronze_cost
10000 * gold_cost + 100 * silver_cost + bronze_cost as total_cost_in_bronze
from items )
, transformed_items2 as (
select *, cast(gold_cost as int) gold_cost,
cast(silver_cost as int) silver_cost,
cast(bronze_cost as int) bronze_cost,

10000 * cast(gold_cost as int) + 100 * cast(silver_cost as int) + cast(bronze_cost as int) as total_cost_in_bronze
from items )
, joined_items_by_acc as
(select * from 
accounts_items ai 
left join
transformed_items i on ai.item_id = i.id
)
, agg_acc_items as
(select account_id, 
count(item_id) as item_cnt, 
sum(gold_cost) as sum_gold_cost,
sum(silver_cost) as sum_silver_cost,
sum(bronze_cost) as sum_bronze_cost
from joined_items_by_acc
group by 1
order by sum(total_cost_in_bronze) desc
)
, acc_items_total_cost as
(select account_id, item_cnt as items, concat(sum_gold_cost,' gold', sum_silver_cost, ' silver', sum_bronze_cost,' bronze') as total_cost from agg_acc_items)
, final_table as (
select a.username, aitc.items, aitc.total_cost 
from acc_items_total_cost aitc
left join accounts a on aitc.account_id = a.id
)
select * from final_table

with transformed_calls as (
select *, dayofweek(dt) as weekday, 
case when dayofweek(dt) in (1,7) then 'weekend' else 'workweek' end as weektime
from calls )
, transformed_calls_agg as (
select contact_id, type, 
sum(case when weektime = 'workweek' then duration end) as workweek_duration,
sum(case when weektime = 'workweek' then 1 end) as workweek_calls,
sum(case when weektime = 'weekend' then duration end) as weekend_duration,
sum(case when weektime = 'weekend' then 1 end) as weekend_calls
from transformed_calls
group by 1,2
order by 1,2
)
, transformed_calls_agg_to_time as
(
select contact_id, type, 
FLOOR(workweek_duration / 3600) as workweek_hours,
FLOOR((workweek_duration - (FLOOR(workweek_duration / 3600) * 3600)) / 60) as workweek_minutes,
workweek_duration - (FLOOR(workweek_duration / 3600) * 3600) - ((FLOOR((workweek_duration - (FLOOR(workweek_duration / 3600) * 3600)) / 60)) * 60) as workweek_seconds,
workweek_calls,

FLOOR(weekend_duration / 3600) as weekend_hours,
FLOOR((weekend_duration - (FLOOR(weekend_duration / 3600) * 3600)) / 60) as weekend_minutes,
weekend_duration - (FLOOR(weekend_duration / 3600) * 3600) - ((FLOOR((weekend_duration - (FLOOR(workweek_duration / 3600) * 3600)) / 60)) * 60) as weekend_seconds,
weekend_calls

FROM 
transformed_calls_agg
)
, transformed_call_to_time_1column as
(select contact_id, type, 
concat(workweek_hours,':',workweek_minutes,':',workweek_seconds,' (',workweek_calls,')') as workweek
concat(weekend_hours,':',weekend_minutes,':',weekend_seconds,' (',workweek_calls,')') as weekend
from 
transformed_calls_agg_to_time
)
, transformed_contacts as
(select *, concat(first_name,' ',last_name) as full_name from contacts)
, final_table as
(
select tc.full_name, tc.phone, tct.type, tct.workweek, tct.weekend
from transformed_call_to_time_1column tct
left join 
transformed_contacts tc on tct.contact_id = tc.id
)
select * from final_table

````python
import pandas as pd
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
````

````python
career = []
original = ['IT','sale','education','comm','pimp']
for i in range(10):
    ls2 = random.sample(original, len(original))
    career.extend(ls2)
    
age = [random.randint(20, 60) for _ in range(50)]
mar = [random.randint(0, 2) for _ in range(50)]
edu = [random.randint(0, 4) for _ in range(50)]
monthlyincome = [random.randint(500, 5000) for _ in range(50)]
contact_type = [random.randint(0, 3) for _ in range(50)]
durations = [random.randint(0, 3000) for _ in range(50)]
contacted_current_campaign = [random.randint(0, 30) for _ in range(50)]
contacted_previous_campaign = [random.randint(0, 30) for _ in range(50)]
did_buy_previous_campaign = [random.randint(0, 1) for _ in range(50)]
did_buy_current_campaign = [random.randint(0, 1) for _ in range(50)]
````

````python
df_fake = pd.DataFrame(
    {'age': age,
     'career': career,
     'marital_status': mar,
     'education_type': edu,
     'monthly_income': monthlyincome,
     'contact_information_type': contact_type,
     'durations': durations,
     'contacted_current_campaign': contacted_current_campaign,
     'contacted_previous_campaign': contacted_previous_campaign,
     'did_buy_previous_campaign': did_buy_previous_campaign,
     'did_buy_current_campaign': did_buy_current_campaign,

    })
````

````python
df_fake
````

### Which job type has the highest mean monthly income?


````python
df_fake.groupby('career').mean()['monthly_income'].sort_values(ascending = False)
````

### Which job type was targeted the most in the previous campaign? Targeting is defined as the mean number of times the target customer was contacted.

````python
df_fake.groupby('career').mean()['contacted_previous_campaign'].sort_values(ascending = False)
````

What is the best reachable contact information type in the current campaign? The best reachable contact information type is defined as the one that was used to contact the target customers with the highest mean number.

````python
df_fake.groupby('contact_information_type').mean()['contacted_current_campaign'].sort_values(ascending = False)
````

For customers contacted more than 30 times, is there a relationship between the mean number of times a target customer was contacted in the previous campaign and the customer's age?

````python
df_fake['total_contacted'] = df_fake['contacted_current_campaign'] + df_fake['contacted_previous_campaign']
df_interest = df_fake[df_fake['total_contacted'] >= 30]
df_interest[['age','contacted_previous_campaign']].plot.scatter(x='age',y='contacted_previous_campaign',c='DarkBlue')


corr, _ = pearsonr(df_interest['age'].to_list(), df_interest['contacted_previous_campaign'].to_list())
print('Pearsons correlation: %.3f' % corr)
````
